```python
def download_data(browser,
                    start_year, start_month, start_day,
                    end_year, end_month, end_day,
                    publisher_option_text='',
                    product_option_text='',
                    catagory_option_text='',
                    tracking_option_text='',
                    group_option_text='',
                    ad_option_text='',
                    download_folder = 'C:/Users/RishiRaj/Downloads',
                    new_location= 'C:/Users/RishiRaj/OneDrive - Zenon/Desktop/temp',
                    ):

    def enter_date(browser,start_year, start_month, start_day, end_year, end_month, end_day):
        global body
        date_range_button = WebDriverWait(browser, 5).until(
        EC.element_to_be_clickable((By.CLASS_NAME, 'sc-fXSgeo')))
        date_range_button.click()

        def set_date_range(browser, start_year, start_month, start_day, end_year, end_month, end_day):
            # Format the start and end dates as strings in 'YYYY-MM-DD' format
            start_date = f"{start_year}-{start_month:02d}-{start_day:02d}"
            end_date = f"{end_year}-{end_month:02d}-{end_day:02d}"
            
            # Wait for the start date input to be clickable and set the start date
            start_date_input = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.NAME, 'startDate'))
            )
            start_date_input.clear()  # Clear any pre-filled value
            start_date_input.send_keys(start_date)
            
            # Assuming the end date input name is 'endDate' (Update accordingly if different)
            end_date_input = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.NAME, 'endDate'))
            )
            end_date_input.clear()  # Clear any pre-filled value
            end_date_input.send_keys(end_date)

        set_date_range(browser, start_year, start_month, start_day, end_year, end_month, end_day)

        body = browser.find_element(By.TAG_NAME, 'body')
        body.send_keys(Keys.TAB * 6 + Keys.ENTER)

    def enter_publisher(option_text):
        if option_text:
            dropdown_trigger = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, 'publisherSelect')))
            dropdown_trigger.click()
            # print(browser.page_source.replace('><','>\n<'))

            all_publishers = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, "react-select-4-option-0")))
            all_publishers.click()
            dropdown_trigger.click()

            desired_option = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.XPATH, f"//div[@class=' css-y0wvtc' and text()='{option_text}']")))
            desired_option.click()
    
    def enter_product(option_text):
        if option_text:
            dropdown_trigger = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, 'productSelect')))
            dropdown_trigger.click()
            # print(browser.page_source.replace('><','>\n<'))

            all = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, "react-select-5-option-0")))
            all.click()
            dropdown_trigger.click()

            desired_option = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.XPATH, f"//div[@class=' css-y0wvtc' and text()='{option_text}']")))
            desired_option.click()
    
    def enter_category(option_text):
            if option_text:
                dropdown_trigger = WebDriverWait(browser, 5).until(
                    EC.element_to_be_clickable((By.ID, 'categorySelect')))
                dropdown_trigger.click()
                # print(browser.page_source.replace('><','>\n<'))

                all = WebDriverWait(browser, 5).until(
                    EC.element_to_be_clickable((By.ID, "react-select-6-option-0")))
                all.click()
                dropdown_trigger.click()

                desired_option = WebDriverWait(browser, 5).until(
                    EC.element_to_be_clickable((By.XPATH, f"//div[@class=' css-y0wvtc' and text()='{option_text}']")))
                desired_option.click()

    def enter_tracking(option_text):
        if option_text:
            dropdown_trigger = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, 'trackingProfileSelect')))
            dropdown_trigger.click()
            # print(browser.page_source.replace('><','>\n<'))

            all = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, "react-select-7-option-0")))
            all.click()
            dropdown_trigger.click()

            desired_option = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.XPATH, f"//div[@class=' css-y0wvtc' and text()='{option_text}']")))
            desired_option.click()

    def enter_group(group_option_text):
        options_group = {
        "Merchant": "react-select-8-option-0",
        "Publisher (ID)": "react-select-8-option-1",
        "Publisher (ID, Name, URL)": "react-select-8-option-2",
        "Date": "react-select-8-option-3",
        "Month": "react-select-8-option-4",
        "Ad (ID)": "react-select-8-option-5",
        "Tracking Profile (ID)": "react-select-8-option-6",
        "Product": "react-select-8-option-7",
        "Product Category": "react-select-8-option-8",
        "Ad Campaign": "react-select-8-option-9"}
        if group_option_text:
            group_dropdown_trigger = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, 'groupSelect')))
            group_dropdown_trigger.click()
            #print(browser.page_source.replace('><','>\n<'))

            group_desired_option = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.ID, options_group[group_option_text])))
            group_desired_option.click()    

    def enter_ad(ad_option_text):
        global ad_dropdown_trigger
        ad_dropdown_trigger = WebDriverWait(browser, 5).until(
            EC.element_to_be_clickable((By.ID, 'adIdSelect')))
        ad_dropdown_trigger.click()
        # print(browser.page_source.replace('><','>\n<'))
        all_ads = WebDriverWait(browser, 5).until(
            EC.element_to_be_clickable((By.ID, "react-select-11-option-0")))
        all_ads.click()
        ad_dropdown_trigger.click()

        if ad_option_text:
            ad_desired_option = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.XPATH, f"//div[@class=' css-y0wvtc' and text()='{ad_option_text}']")))
            ad_desired_option.click()
        #ad_dropdown_trigger.click()
    
    def download():
        global body, ad_dropdown_trigger, success
        files_before = os.listdir(download_folder)
        ad_dropdown_trigger.click()
        while True:
            body.send_keys(Keys.TAB)
            if "Click generate reports to get started" in browser.page_source:
                if 'Customize Columns' in browser.switch_to.active_element.get_attribute('outerHTML'):
                    body.send_keys(Keys.TAB + Keys.ENTER)
                    break
        
        def try_download():
            global success
            if ("Page 1 of" in browser.page_source):
                ad_dropdown_trigger.click()
                while True:
                    body.send_keys(Keys.TAB)
                    if "react-select-12-input" in browser.switch_to.active_element.get_attribute('outerHTML'):
                        body.send_keys(Keys.TAB + Keys.ENTER)
                        success=True
                        break

        def check_download():
            files_after = os.listdir(download_folder)
            new_files = [f for f in files_after if f not in files_before]
            if new_files:
                print("Download has started. New file detected:", new_files)
                return True
            else: return False

        try_download()
        while True:
            if check_download() or "We couldn't find what you're looking for. Please try it again." in browser.page_source:
                break
            else: try_download()   
       
    def wait_for_download_completion(timeout=300):
        end_time = time.time() + timeout
        while True:
            if time.time() > end_time:
                raise Exception("Timeout waiting for download to complete")
            crdownload_files = [f for f in os.listdir(download_folder) if f.endswith('.crdownload')]
            if not crdownload_files:
                break

    def rename_and_move_file(new_name):
        # Find the most recently downloaded file
        files = os.listdir(download_folder)
        full_paths = [os.path.join(download_folder, f) for f in files]
        latest_file = max(full_paths, key=os.path.getctime)

        # Determine the new file path
        _, extension = os.path.splitext(latest_file)
        new_file_path = os.path.join(new_location, f"{new_name}{extension}")

        # Move and rename the file
        shutil.move(latest_file, new_file_path)
        print(f"File moved and renamed to: {new_file_path}")
    
    global success
    success = False
    
    enter_date(browser, start_year, start_month, start_day, end_year, end_month, end_day)
    enter_publisher(publisher_option_text)
    enter_product(product_option_text)
    enter_category(catagory_option_text)
    enter_tracking(tracking_option_text)
    enter_group(group_option_text)
    enter_ad(ad_option_text)
    download()
    if success:
        all_var= [i.replace(' ', '') for i in [publisher_option_text,product_option_text,catagory_option_text,tracking_option_text,group_option_text,ad_option_text] if i]
        print(all_var)
        wait_for_download_completion()
        time.sleep(0.2)  
        date_range = f"{start_year}{start_month:02d}{start_day:02d}-{end_year}{end_month:02d}{end_day:02d}"
        new_name = f"{date_range}"
        for i in all_var:
            new_name = f"{new_name}_)(_{i}"
        print(new_name)
        rename_and_move_file(new_name)

def main():
    global browser
    browser.get('https://app.fintelconnect.com/merchant/performance-report')
    toggle_advance = WebDriverWait(browser, 20).until(
        EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'Show Advanced Search')]")))
    toggle_advance.click()

    y = [2024]
    m = list(range(1,13))
    d = list(range(14,31))

    for pub in ["26869 - Taboola"]:
        for day in d:
            start_year, start_month, start_day = 2024,1,day
            end_year, end_month, end_day = 2024,1,day+1
            publisher_option_text = pub
            tracking_option_text = '30299 - Time.com-Personal Finance'
            group_option_text = ''
            ad_option_text = "26650 - High-Yield Savings Premier"

            download_data(browser,start_year, start_month, start_day, end_year, end_month, end_day, publisher_option_text = pub,ad_option_text = ad_option_text, tracking_option_text=tracking_option_text) 
            # clear_trigger = WebDriverWait(browser, 5).until(
            #     EC.element_to_be_clickable((By.CLASS_NAME, 'esLGqj')))
            # clear_trigger.click()

main()
```
